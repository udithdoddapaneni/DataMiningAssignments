## 1d. Python Setup

In this subtask, we will setup an isolated Python environment for the programming assignment. The goal of this task is have a project setup that can keep running for years to come. In a typical enterprise environment, it is not unheard of for an application to be running for years if not decades. A recent exhibit came during the CrowdStrike debacle when SouthWest airlines in USA (a budget carrier like Indigo) was not affected because their system runs on Windows 3.1 which came in 1992. Even if that is an extreme case, it is imperative to make sure that your code runs for a few years.

Most of the common ways in which data science projects are organized are unsuited for this purpose. Typically, students and young data scientists use a global Python and Python packages. So upgrading a library inside one project will have unexpected impact on other projects. Virtual environments is a good start and solves some but not all of the issues. The goal of this sub task is to have a working setup that will ensure that it is isolated and will not be affected by the changes in the rest of your system.

### Recommendation: Avoid Using the Python Installation in Your Operating System 

If you are using Linux or macOS, it is likely that your OS already comes with Python. A rookie mistake is to start using this version for your projects. This version is typically designed for OS tools. So, it is likely that it is using an older version of Python. For example, even though I have the latest version of macOS (Sonoma, as Sequoia is not yet released), the Python version is 3.9 which was released 3-4 years ago whereas Python 3.13 is going to be released in October. 

Using an older version is bad for many reasons. First, you will not have the latest cool functionalities that the vibrant Python community is developing. Second, faster-cpython project has been steadily making the interpreter fast. So you are missing out on potential performance improvement where you do not need to do anything :) Finally, lot of libraries we will use might not even work with old Python versions.

Given that we are doing data science stuff, a good rule-of-thumb is to always use one version behind the latest stable release. For example, if the latest version is 3.X, then use 3.(X-1). The reason is that it takes few months before major DS libraries (PyTorch being the most notorious example) make it work seamlessly for the new version. Of course, if you do not use lot of DS stuff, stick with the latest version. By a quirk of timing, we will be using Python 3.12 even though it is the latest version. The reason is that Python 3.13 will be released in October and 3.12 was released a year before that - so most vendors have had a lot of time to make their libraries work for 3.12 . 

### Recommendation: Install Python With Tools That Support Multiple Versions

Using the latest version is just one part of the puzzle. As Python keeps improving, we will need to use more recent versions. However, it is possible that your old DS projects rely on the old Python versions (this case is surprisingly common). So, when installing Python, do not simply go to [python.org](https://www.python.org/) and run the installer in that website. You should use a more sophisticated tool that allows you to install multiple Python versions. This is a surprisingly complex task to do on your own.

Two most common tools for this are [pyenv](https://github.com/pyenv/pyenv) and [mise](https://mise.jdx.dev/). I regularly use pyenv. In an enterprise environment, I would recommend you to use pyenv. 

In our course, we will take a more complex approach that I think will become standardized in the future. We will using a very powerful tool called [Rye](https://rye.astral.sh/) that can do this and more. 


## 1e. Setting up Isolated Virtual Environments using Rye

Python has a very fragmented development ecosystem. This is in contrast to other languages such as Rust or Golang that has excellent toolchain that comes as part of the language installation itself. There has been some efforts to build "Cargo for Python" and it might take couple of years before it matures. 

[Rye](https://rye.astral.sh/) is the new kid on the block that has become extremely popular even though it is quite new and also experimental. My original plan for the assignment was to use a combination of [pyenv](https://github.com/pyenv/pyenv) and [PDM](https://pdm-project.org/en/latest/). Some of the recent features added to Rye makes it easier to use it as a single tool that can do setting up multiple Python versions and setting up virtual environments. 

It is unlikely that Rye becomes the dominant tool but its design principles (using other tools such as `uv`, `pdm`, `pip-tools` etc instead of building yet another tool) ensures that the eventual tool will reuse most of its ideas. Rye has been adopted by [Astral](https://astral.sh/) that has excellent track record in dev tooling systems including `Ruff` and `uv`.  

Another popular library for virtual environments is [poetry](https://python-poetry.org/). It pioneered many of the techniques that is used by Rye, PDM and Hatch. However, since it came too early, it also uses many non-standard implementations for virtual environments that differs from the accepted PEPs. For example, it uses a different set of keys in pyproject.toml that can affect compatibility with other tools. Do not use it for any new projects.

## 1g. Setting up pipx commands

By default, all the libraries that are needed for a project will be installed within the virtual environment. A good advice is to ensure that all packages are specified in the dependencies and dev-dependencies key of pyproject.toml.

However, there are many other libraries that are not core to your project and you might not want to install them within the project. For example, you might use something like `ruff` or `pylint` but you might want to use it globally instead of duplicating it per project. 

A natural, but incorrect, approach is to use `pip`. The issue is that blindly installing the package using pip will result in some of the same issues discussed above. They are often tied to a fixed Python version. So a library that you installed for Python 3.11 might not be available for Python 3.12. Even worse, installing a library might bump the version of some other dependent library and break a third library. So the modern approach is to use `pipx` that installs a library and all its dependent libraries in a self-contained environment. This allows that library to operate independently and ensures that it will keep working even if you upgrade other libraries. 

We will not use pipx in this assignment but in the future ones. Nevertheless, it is a useful tool in your toolkit. When installing python packages in the future, take a step back and decide whether to use pipx or install it within a virtual environment.
